```python
import string
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
with open('shakespeare.txt','r') as f:
    for line in f:
        for word in line.split():
            word = word.lower()
            word = word.translate(str.maketrans('', '', string.punctuation))
            print(word)
            
```
